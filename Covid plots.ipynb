```python
import matplotlib.dates as mdates

def my_single_axis_plot(xdata,yseries,colorseries,xlabel,ylabel,linelabels,
                        xlabelcolor,ylabelcolor,areaname,filename):
    fig, ax1 = plt.subplots(figsize=(6, 4))
    fs = 20
    lfs = 18
    legendfs = 16

    n = len(yseries)
    ax1.set_xlabel(xlabel,fontsize = fs)
    ax1.set_ylabel(ylabel,fontsize = fs)
    ax1.tick_params(axis='x', labelcolor=xlabelcolor, labelsize = lfs)
    ax1.tick_params(axis='y', labelcolor=ylabelcolor, labelsize = lfs)

    for i in range(n):
        ax1.plot(xdata, yseries[i], color=colorseries[i], label=linelabels[i])
        ax1.legend(prop={'size':legendfs})
    
    ax1.set_xticklabels(xdata, rotation = 45, ha="right",fontsize = lfs)
    myFmt = mdates.DateFormatter('%d %b')
    ax1.xaxis.set_major_formatter(myFmt)

    plt.grid(True,color='k', linestyle='-', linewidth=1, alpha = 0.2)
    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.title(areaname, fontsize = fs)
    plt.show()
    if filename != '':
        fig.savefig("results/" + areaname+"_"+filename+".pdf", bbox_inches='tight')
    
def my_double_axes_plot(xdata,
                        yseries1,colorseries1,
                        yseries2,colorseries2,
                        xlabel,ylabel1,ylabel2,
                        linelabels1,linelabels2,
                        xlabelcolor,ylabelcolor1,ylabelcolor2,
                       areaname,filename):
    
    fig, ax1 = plt.subplots(figsize=(10, 7))
    fs = 20
    lfs = 18
    legendfs = 16
    ax1.set_xlabel(xlabel,fontsize = fs)
    myFmt = mdates.DateFormatter('%d %b')
    ax1.xaxis.set_major_formatter(myFmt)
    
    ax1.set_ylabel(ylabel1, color=ylabelcolor1,fontsize = fs)
    p = []
    for i in range(len(yseries1)):
        ax1.plot(xdata, yseries1[i], color=colorseries1[i], label=linelabels1[i])
        ax1.legend(bbox_to_anchor=(1.2, 1), loc='upper left', prop={'size':legendfs})
    ax1.tick_params(axis='y', labelcolor=color, labelsize = lfs)
    ax1.set_xticklabels(xdata, rotation = 45, ha="right",fontsize = lfs)
    myFmt = mdates.DateFormatter('%d %b')
    ax1.xaxis.set_major_formatter(myFmt)
#     ax1.set_yticklabels(fontsize = fs)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    ax2.set_ylabel(ylabel2, color=ylabelcolor2,fontsize = fs)  # we already handled the x-label with ax1
    for i in range(len(yseries2)):
        ax2.plot(xdata, yseries2[i], color=colorseries2[i], label=linelabels2[i])
        ax2.legend(bbox_to_anchor=(1.2, 0.4), loc='lower left', prop={'size':legendfs})
    ax2.tick_params(axis='y', labelcolor=ylabelcolor2, labelsize = lfs)
#     ax2.set_yticklabels(fontsize = fs)
    ax2.xaxis.set_major_formatter(myFmt)

#     fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.grid(True,color='k', linestyle='-', linewidth=1, alpha = 0.2)
    plt.title(areaname, fontsize = fs)
    plt.show()
    if filename != '':
        fig.savefig("results/" + areaname+"_"+filename+".pdf", bbox_inches='tight')

def my_double_axes_plot_inside_legend(xdata,
                        yseries1,colorseries1,
                        yseries2,colorseries2,
                        xlabel,ylabel1,ylabel2,
                        linelabels1,linelabels2,
                        xlabelcolor,ylabelcolor1,ylabelcolor2,
                       areaname,filename):
    
    fig, ax1 = plt.subplots(figsize=(10, 7))
    fs = 30
    lfs = 24
    legendfs = 20
    ax1.set_xlabel(xlabel,fontsize = fs)
    myFmt = mdates.DateFormatter('%d %b')
    ax1.xaxis.set_major_formatter(myFmt)
    
    ax1.set_ylabel(ylabel1, color=ylabelcolor1,fontsize = fs)
    p = []
    for i in range(len(yseries1)):
        ax1.plot(xdata, yseries1[i], color=colorseries1[i], label=linelabels1[i])
        ax1.legend(bbox_to_anchor=(0, 0.90), loc='upper left', prop={'size':legendfs})
    ax1.tick_params(axis='y', labelcolor=color, labelsize = lfs)
    ax1.set_xticklabels(xdata, rotation = 45, ha="right",fontsize = lfs)
    myFmt = mdates.DateFormatter('%d %b')
    ax1.xaxis.set_major_formatter(myFmt)
#     ax1.set_yticklabels(fontsize = fs)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    ax2.set_ylabel(ylabel2, color=ylabelcolor2,fontsize = fs)  # we already handled the x-label with ax1
    for i in range(len(yseries2)):
        ax2.plot(xdata, yseries2[i], color=colorseries2[i], label=linelabels2[i])
        ax2.legend(bbox_to_anchor=(0, 0.85), loc='lower left', prop={'size':legendfs})
    ax2.tick_params(axis='y', labelcolor=ylabelcolor2, labelsize = lfs)
#     ax2.set_yticklabels(fontsize = fs)
    ax2.xaxis.set_major_formatter(myFmt)

#     fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.grid(True,color='k', linestyle='-', linewidth=1, alpha = 0.2)
    plt.title(areaname, fontsize = fs)
    plt.show()
    if filename != '':
        fig.savefig("results/" + areaname+"_"+filename+".pdf", bbox_inches='tight')
        
def my_single_axis_wide_plot(xdata,yseries,colorseries,xlabel,ylabel,linelabels,
                        xlabelcolor,ylabelcolor,areaname,filename):
    fig, ax1 = plt.subplots(figsize=(10, 7))
    fs = 30
    lfs = 24
    legendfs = 20

    n = len(yseries)
    ax1.set_xlabel(xlabel,fontsize = fs)
    ax1.set_ylabel(ylabel,fontsize = fs)
    ax1.tick_params(axis='x', labelcolor=xlabelcolor, labelsize = lfs)
    ax1.tick_params(axis='y', labelcolor=ylabelcolor, labelsize = lfs)

    for i in range(n):
        ax1.plot(xdata, yseries[i], color=colorseries[i], label=linelabels[i])
        ax1.legend(loc = "upper right", prop={'size':legendfs})
    
    ax1.set_xticklabels(xdata, rotation = 45, ha="right",fontsize = lfs)
    myFmt = mdates.DateFormatter('%d %b')
    ax1.xaxis.set_major_formatter(myFmt)

    plt.grid(True,color='k', linestyle='-', linewidth=1, alpha = 0.2)
    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.title(areaname, fontsize = fs)
    plt.show()
    if filename != '':
        fig.savefig("results/" + areaname+"_"+filename+".pdf", bbox_inches='tight')

```

```python
def read_dataset(areaname):
    od = pd.read_excel('covid_processed.xlsx',sheet_name=areaname, skiprows=0)
    ndates = od['Date']
    dates = []
    for i in range(len(ndates)):
        dates.append(ndates[i].date())

    return (od, dates)

od, dates = read_dataset('Odisha')
```

```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure
from bokeh.io import output_file, show, output_notebook
from bokeh.models import CustomJS
from bokeh.models.widgets import CheckboxGroup
from bokeh.layouts import row
from bokeh.palettes import Viridis5
from bokeh.models.annotations import Title, Legend

# df = pd.read_csv("TSLA.csv")

# df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
df = od
p = figure(x_axis_type='datetime', plot_width=800)
passenger_type = 'Train passengers'
aline = p.line(od['Date'], od[passenger_type], line_width=2, color=Viridis5[0])
bline = p.line(Cd['Date'], Cd[passenger_type], line_width=2, color=Viridis5[1])
cline = p.line(Gd['Date'], Gd[passenger_type], line_width=2, color=Viridis5[2])
dline = p.line(Sd['Date'], Sd[passenger_type], line_width=2, color=Viridis5[3])
eline = p.line(Kd['Date'], Kd[passenger_type], line_width=2, color=Viridis5[4])

p.yaxis.axis_label = passenger_type
p.xaxis.axis_label = 'Dates'

legend = Legend(items=[
    ("Odisha",   [aline]),
    ("Cuttack", [bline]),
    ("Ganjam", [cline]),
    ("Sundargarh", [dline]),
    ("Khordha", [eline])
], location=(0, 350))

t = Title()
t.text = passenger_type
p.title = t

p.add_layout(legend, 'left')

checkboxes = CheckboxGroup(labels=list(['Odisha', 'Cuttack', 'Ganjam', 
                          'Sundargarh', 'Khordha']), active=[0, 1, 2, 3, 4])
callback = CustomJS(code="""aline.visible = false; // aline and etc.. are 
                            bline.visible = false; // passed in from args
                            cline.visible = false; 
                            dline.visible = false;
                            eline.visible = false;
                            // cb_obj is injected in thanks to the callback
                            if (cb_obj.active.includes(0)){aline.visible = true;} 
                                // 0 index box is aline
                            if (cb_obj.active.includes(1)){bline.visible = true;} 
                                // 1 index box is bline
                            if (cb_obj.active.includes(2)){cline.visible = true;} 
                                // 2 index box is cline etc...
                            if (cb_obj.active.includes(3)){dline.visible = true;}
                            if (cb_obj.active.includes(4)){eline.visible = true;}""",
                    args={'aline': aline, 'bline': bline, 'cline': cline, 'dline': dline, 'eline':eline})
checkboxes.js_on_click(callback)
output_file(passenger_type+'comparison.html')
show(row(p, checkboxes))
```

### Show scatter plot of incoming migrants versus heterogeneous risk levels
